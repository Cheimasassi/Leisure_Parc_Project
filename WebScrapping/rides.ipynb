```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Start a WebDriver session
driver = webdriver.Chrome()

# Load the webpage
driver.get("https://www.worldsoffun.com/rides-experiences")

# Get the page source
page_source = driver.page_source

# Close the WebDriver session
driver.quit()

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find all ride elements
rides = soup.find_all(class_="cf-filtered-poi-item")

# Create empty lists to store data
ride_names = []
attraction_types = []
min_heights = []

# Iterate over each ride
for ride in rides:
    ride_name = ride.find(class_='cf-filtered-poi-item__title').get_text(strip=True)
    ride_names.append(ride_name)
    
    properties = ride.find_all(class_='cf-filtered-poi-properties__prop')
    
    attraction_type = "Not Specified"
    min_height = "Not Specified"
    
    for prop in properties:
        key = prop.find(class_='cf-filtered-poi-properties__key').get_text(strip=True)
        value = prop.find(class_='cf-filtered-poi-properties__value').get_text(strip=True)
        
        if key == "Attraction Type:":
            attraction_type = value
        elif key == "Minimum Height:":
            min_height = value
            
    attraction_types.append(attraction_type)
    min_heights.append(min_height)

# Create a DataFrame
df = pd.DataFrame({
    'Ride Name': ride_names,
    'Attraction Type': attraction_types,
    'Minimum Height': min_heights
})

# Save the DataFrame to a CSV file
df.to_csv('rides.csv', index=False)

print(df.head())  # Print the first few rows of the DataFrame to verify the data
```
