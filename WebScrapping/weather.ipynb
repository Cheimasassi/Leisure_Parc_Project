```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Start a WebDriver session
driver = webdriver.Chrome()

# Load the webpage
driver.get("https://www.historique-meteo.net/afrique/tunisie/tunis/2023/11/?fbclid=IwAR2A5Z8xx4udJWK34Ibdlvmo-e159ti9Q43eMCHCC_0p")
# Get the page source
page_source = driver.page_source

# Close the WebDriver session

soup = BeautifulSoup(page_source, "html.parser")

# Find all <tr> tags
rows = soup.find_all("tr")

# Initialize lists to store extracted data
jour_list = []
temperatures_list = []
precipitations_list = []
avis_list = []

# Extract data from each row
for row in rows:
    cols = row.find_all("td")
    if len(cols) >= 4:
        jour = cols[0].text.strip()
        data_str = cols[2].text.strip()
        if "Températures :" in data_str:
            temperatures = data_str.split("Températures : ")[1].split("<br>")[0]
        else:
            temperatures = None
        if "Précipitations :" in data_str:
            precipitations = data_str.split("Précipitations : ")[1]
        else:
            precipitations = None
        avis = cols[3].text.strip()
        
        jour_list.append(jour)
        temperatures_list.append(temperatures)
        precipitations_list.append(precipitations)
        avis_list.append(avis)


# Create a pandas DataFrame
data = {
    "Jour": jour_list,
    "Températures": temperatures_list,
    "Précipitations": precipitations_list,
    "Avis sur la météo": avis_list
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("weather_data_november.csv", index=False)

# Close the Selenium webdriver
driver.quit()
```

```python
import pandas as pd
import re

# Charger le fichier CSV dans un DataFrame
df = pd.read_csv('weather_data_august')

# Définir la fonction de nettoyage
def nettoyer_temperature(texte):
    # Utiliser une expression régulière pour supprimer la partie "Précipitations : xxx"
    texte_nettoye = re.sub(r'Précipitations : \d+[^a-zA-Z0-9]', '', str(texte))
    
    # Remplacer la lettre "Ã" par une chaîne vide
    texte_nettoye = texte_nettoye.replace('Ã', '')
    
    return texte_nettoye

# Appliquer la fonction de nettoyage à la colonne appropriée (par exemple, 'temperature_col')
df['Temperature'] = df['Temperature'].apply(nettoyer_temperature)

# Enregistrer le DataFrame nettoyé dans un nouveau fichier CSV
df.to_csv('nouveau_fichier.csv', index=False)
```
