```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Start a WebDriver session
driver = webdriver.Chrome()

# Load the webpage
driver.get("https://www.silverwoodthemepark.com/tickets/")

# Get the page source
page_source = driver.page_source

# Close the WebDriver session
driver.quit()

soup = BeautifulSoup(page_source, 'html.parser')

# Extracting data from tables
tables = soup.find_all('table')

# List to hold extracted data
data = []

# Loop through tables
for table in tables:
    # Determine if it's online or gate
    ticket_type = "Online" if "Online" in table.th.text else "Gate"
    
    # Extract data from rows
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all(['td', 'th'])
        cols = [col.text.strip() for col in cols]
        if len(cols) > 1:
            ticket_category = cols[0].split('\n')[0]
            prices = cols[1:]
            data.append([ticket_type, ticket_category] + prices)

# Convert data to DataFrame
df = pd.DataFrame(data, columns=['Type', 'Category', '1-day Admission', '2-day Admission', '3-day Admission', 'After 4pm Admission'])

# Save DataFrame to CSV
df.to_csv('ticket_prices.csv', index=False)

print("CSV file created successfully.")
```
