```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# Start a WebDriver session
driver = webdriver.Chrome()

# Load the webpage
driver.get("https://www.silverwoodthemepark.com/events/")

# Get the page source
page_source = driver.page_source

# Close the WebDriver session
driver.quit()

# Parse the HTML using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find all event sections
event_sections = soup.find_all("div", class_="event")

event_data = []

for event in event_sections:
    # Extract date and name
    date_element = event.find("strong")
    date = date_element.text if date_element else None
    
    name_element = event.find("h1")
    name = name_element.text if name_element else None
    
    # Extract description
    description_element = event.find("p")
    description = description_element.text if description_element else None
    
    # Append data to the list
    event_data.append({
        "Date": date,
        "Name": name,
        "Description": description
    })

# Create a DataFrame from the list
df = pd.DataFrame(event_data)

# Save DataFrame to a CSV file
df.to_csv("events_data.csv", index=False)
```
