### <span style="color:blue">Relation Between Orders and Orderproducts tables</span>

```python
orders = pd.read_csv('E:/ProjetBI/data/orders.csv')
orderproduct = pd.read_csv('E:/ProjetBI/data/orderproducts.csv')
```

```python
orders['day'] = pd.to_datetime(orders['access_date']).dt.day_name()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=orderproduct["product_id"], y=orders["day"], hue=orderproduct["quantity"], palette='coolwarm', s=100, alpha=0.7)
plt.xlabel('Product ID')
plt.ylabel('Day')
plt.title('Distribution of Product ID by Day')
plt.xticks(orderproduct["product_id"].unique())
plt.yticks(orders["day"].unique())
plt.legend(title='Quantity', loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

```

The scatter plot above showcases the distribution of products by day, providing insights into the most and least demanded products on a daily basis. Notably, product IDs 8 and 18 emerge as the most frequently ordered products, while product ID 4 appears to be the least popular.

This visualization allows businesses to identify patterns and trends in product demand, enabling them to optimize inventory management and make informed decisions regarding product offerings and marketing strategies. By analyzing this data, businesses can ensure they meet customer preferences and maximize sales opportunities.


```python
merged_df = pd.merge(orderproduct, orders, left_on='order_id', right_on='id')
display(merged_df.head())
plt.figure(figsize=(12, 12))
orderproduct_grouped = merged_df.groupby('day')['quantity'].sum()
orders_grouped = merged_df.groupby('day')['total_y'].sum()
orders_socks=merged_df.groupby('day')['number_sock'].sum()

plt.subplot(2, 2, 1)
orderproduct_grouped.plot(kind='bar', color='skyblue', alpha=0.7)
plt.xlabel('Day')
plt.ylabel('Total Quantity')
plt.title('Total Quantity by Day')

plt.subplot(2, 2, 2)
orders_grouped.plot(kind='bar', color='orange', alpha=0.7)
plt.xlabel('Day')
plt.ylabel('Total Amount')
plt.title('Total by Day')

plt.subplot(2, 2, 3)
orders_socks.plot(kind='bar', color='red', alpha=0.7)
plt.xlabel('Day')
plt.ylabel('Total Socks')
plt.title('Total Socks by Day')

plt.tight_layout()
plt.show()
```

The analysis conducted in the previous cells provides valuable insights into the distribution and trends of product orders. By examining the quantity of products ordered on each day of the week, as well as the total amount and the total number of socks ordered, we can draw meaningful conclusions about the park's performance.

The bar plots reveal that Saturday and Sunday exhibit higher product orders compared to the rest of the week. This information suggests that weekends are the peak days for the park, attracting a larger number of visitors and generating higher sales. Conversely, the weekdays show lower product orders, indicating lower visitor traffic during these days.

These findings can be instrumental in optimizing park operations and marketing strategies. By focusing resources and promotional efforts on weekends, businesses can maximize revenue and enhance customer experiences. Additionally, understanding the demand patterns throughout the week can aid in effective inventory management and staffing decisions.
